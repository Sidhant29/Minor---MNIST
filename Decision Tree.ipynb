# Loading the Dataset and Reshaping it

```
(x_train, y_train), (x_test, y_test) = mnist.load_data()
X_ = x_train
Y_ = y_train
X__ = x_test
Y__ = y_test
print (x_train.shape)
print (y_train.shape)
print (x_test.shape)
print (y_test.shape)
```

```
x_train = x_train.reshape(-1,784)
x_test = x_test.reshape(-1,784)
print (x_train.shape)
print (y_train.shape)
print (x_test.shape)
print (y_test.shape)
```

```
X = x_train/255.0
y = y_train
X = np.array(X)
y = np.array(y)
```

# Fitting the model

```
d0 = datetime.datetime.now()
print ("Time before training : " , str(d0))
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
d1 = datetime.datetime.now()
print ("Time after training : " , str(d0))
print ("Training : " , str(d1-d0))
```

# Plotting the Tree

```
import pydotplus
from sklearn.externals.six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
dot_data = StringIO()
export_graphviz(clf,out_file=dot_data,filled=True,rounded=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

```
X = X_.reshape(-1,784)
Y = Y_
split = int(X.shape[0]*0.80)
x_train = X[:split]
y_train = Y[:split]
x_val = X[split:]
y_val = Y[split:]
x_test = X__.reshape(-1,784)
y_test = Y__

print (split)
print (X.shape,Y.shape)
print (x_train.shape,y_train.shape)
print (x_val.shape,y_val.shape)
print (x_test.shape,y_test.shape)
```

# Accuracy

```
print ("Training Data")
print ("Number of samples : ", str(x_train.shape[0]))
y_pred = clf.predict(x_train)
print('Training Accuracy = ' , str(metrics.accuracy_score(y_train, y_pred)))

print ('#'*15)

print ("Validation Data")
print ("Number of samples : ", str(x_val.shape[0]))
y_pred = clf.predict(x_val)
print("Validation Accuracy = ", str(metrics.accuracy_score(y_val, y_pred)))

print ('#'*15)

print ("Testing Data")
print ("Number of samples : ", str(x_test.shape[0]))
y_pred = clf.predict(x_test)
print("Testing Accuracy = ", str(metrics.accuracy_score(y_test, y_pred)))
```
