```python
import pandas as pd
from keras.datasets import mnist
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import datetime
```

```python
(x,y),(x_,y_) = mnist.load_data()
x = x.reshape(-1,784)
x_ = x_.reshape(-1,784)

print (x.shape,y.shape)
print (x_.shape,y_.shape)
```

```python
split = int(x.shape[0]*0.80)

x_train = x[:split]
y_train = y[:split]
x_val = x[split:]
y_val = y[split:]
x_test = x_ 
y_test = y_

print (x_train.shape, x_val.shape, x_test.shape)
print (y_train.shape, y_val.shape, y_test.shape)
```

```python
logisticRegr = LogisticRegression()
```

```python
d0 = datetime.datetime.now()
print (str(d0))
```

```python
logisticRegr.fit(x, y)
```

```python
d1 = datetime.datetime.now()
print (str(d1))
```

```python
print ("Total training Time : ", str(d1-d0))
```

```python
logisticRegr.predict(x_test[0:20])
```

```python

score1 = logisticRegr.score(x_train, y_train)
score2 = logisticRegr.score(x_val, y_val)
score3 = logisticRegr.score(x_test, y_test)


print ("For Training Data")
print ("     Accuracy = "+str(score1))
print ("For Testing Data")
print ("     Accuracy = "+str(score3) )
print ("For Validation Data")
print ("     Accuracy = "+str(score2) )
```
